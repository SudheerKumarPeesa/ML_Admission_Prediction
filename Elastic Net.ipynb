# Step-3:Data Preprocessing

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
x=df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP',
       'LOR ', 'CGPA', 'Research']]
y=df['Chance of Admit ']
```

```python
# model - Polynomial Regression

from  sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=8)
```

# Step-4: Modelling

```python
#modelling
from sklearn.linear_model import ElasticNet

enr_base=ElasticNet()
enr_base.fit(x_train,y_train)

#prediction

train_predictions = enr_base.predict(x_train)

test_predictions = enr_base.predict(x_test)

# evaluation 

print("Train R2:",enr_base.score(x_train,y_train))

print("Test R2:",enr_base.score(x_test,y_test))

from sklearn.model_selection import cross_val_score
print("CV Score:", cross_val_score(enr_base, x,y, cv=5).mean())

      
```

# Applying Hyperparameter tuning for Lasso Regression

```python

#model
from sklearn.model_selection import GridSearchCV
estimator = ElasticNet()

#parameters  & values
param_grid = {"alpha":[1,0.2,3,5,4,6,8,9,7,2],'l1_ratio':[0.1,0.3,0.4,0.2,0.5,0.6,0.8,0.9,0.7,1]}
# identify the best value of the parameter within given for the given data

model_hp=GridSearchCV(estimator,param_grid,cv=5,scoring='neg_mean_squared_error')

model_hp.fit(x_train,y_train)
model_hp.best_params_
```

# Rebuilt Lasso Model using best hyperparameters

```python
#modelling
enr_best=ElasticNet(alpha=0.2,l1_ratio=0.1)
enr_best.fit(x_train,y_train)
print("Intercept",enr_best.intercept_)
print("coefficient",enr_best.coef_)

#prediction

train_predictions = enr_base.predict(x_train)

test_predictions = enr_base.predict(x_test)

# evaluation 

print("Train R2:",enr_base.score(x_train,y_train))

print("Test R2:",enr_base.score(x_test,y_test))

from sklearn.model_selection import cross_val_score
print("CV Score:", cross_val_score(enr_base, x,y, cv=5).mean())

      
```
