```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel(r'C:\Users\Sudheerkumar Peesa\Machine Learning Projects\Admission_Predict Data   ( REGRESSION PROJECT )\cleaned_data_Admission.xlsx')
df
```

```python
df=df.drop(columns=["Unnamed: 0"])
```

```python
df
```

```python
x=df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP',
       'LOR ', 'CGPA', 'Research']]
y=df['Chance of Admit ']
```

# Train-Test Split

```python
# model - Polynomial Regression

from  sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=8)
```

```python
x_train
```

```python
x_test
```

# Modellingand Evaluation
Applying Hyperparameter tuning for Lasso Regression

```python
from sklearn.model_selection import GridSearchCV


#model
from sklearn.linear_model import Lasso
estimator = Lasso()

#parameters  & values
param_grid = {"alpha":list(range(1,50))}
# identify the best value of the parameter within given for the given data

model_hp=GridSearchCV(estimator,param_grid,cv=5,scoring='r2')

model_hp.fit(x_train,y_train)
model_hp.best_params_
```

Build Lasso Model best hyperparameters

```python
#modelling

from sklearn.linear_model import Lasso
lasso_best = Lasso(alpha=1)
lasso_best.fit(x_train,y_train)

print("Intercept",lasso_best.intercept_)
print("coefficient",lasso_best.coef_)


#prediction & Evaluation on train data 

ypred_train =lasso_best.predict(x_train)

from sklearn.metrics import r2_score 
print("Train R2:",r2_score(y_train, ypred_train))

from sklearn.model_selection import cross_val_score
print("CV Score:", cross_val_score(lasso_best, x_train,y_train, cv=5).mean())

#prediction & Evaluation on test data

ypred_test=lasso_best.predict(x_test)
print("Test R2",r2_score(y_test,ypred_test))
```

```python
df.drop(columns=['TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
       'Research'])
```

```python
df.drop(columns=['TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
       'Research'],inplace=True)

x=df.drop(columns=['Chance of Admit '])
y=df['Chance of Admit ']


from  sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=8)

from sklearn.linear_model import Lasso
lasso_best = Lasso(alpha=1)
lasso_best.fit(x_train,y_train)

print("Intercept",lasso_best.intercept_)
print("coefficient",lasso_best.coef_)


#prediction & Evaluation on train data 

ypred_train =lasso_best.predict(x_train)

from sklearn.metrics import r2_score 
print("Train R2:",r2_score(y_train, ypred_train))

from sklearn.model_selection import cross_val_score
print("CV Score:", cross_val_score(lasso_best, x_train,y_train, cv=5).mean())

#prediction & Evaluation on test data

ypred_test=lasso_best.predict(x_test)
print("Test R2",r2_score(y_test,ypred_test))
```
